# Plotting degree of africanization in Ithaca bees

```python
from reportlab.lib.units import cm
from reportlab.lib import colors
from Bio.SeqFeature import SeqFeature, FeatureLocation
from Bio.Graphics import BasicChromosome
from collections import defaultdict
import pandas as pd
import numpy as np
import sets
```

```python
import MySQLdb
db = MySQLdb.connect('ecoevo.unit.oist.jp','bees',"apis","bees")
cursor = db.cursor()
cursor.connection.autocommit(True)
```

Load data from SQL, subtracting modern allele frequencies from modern

```python
sql = """
CREATE TEMPORARY TABLE temp.modern (
SELECT ahb_fst.chrom, ahb_fst.pos,pop,freq FROM (SELECT chrom, pos FROM freqs WHERE chrom = "Group16" AND pos < 100000 GROUP BY chrom,pos HAVING COUNT(pop) = 2 AND MAX(freq)>0.1 ) AS both_freqs
JOIN ahb_fst
ON ahb_fst.chrom = both_freqs.chrom AND ahb_fst.pos = both_freqs.pos
JOIN freqs
ON  both_freqs.chrom = freqs.chrom  AND both_freqs.pos = freqs.pos AND pop = "modern"
ORDER BY chrom,pos
);
"""
cursor.execute(sql)
sql = """
CREATE TEMPORARY TABLE temp.old (
SELECT ahb_fst.chrom, ahb_fst.pos,pop,freq FROM (SELECT chrom, pos FROM freqs WHERE chrom = "Group16" AND pos < 100000 GROUP BY chrom,pos HAVING COUNT(pop) = 2 AND MAX(freq)>0.1 ) AS both_freqs
JOIN ahb_fst
ON ahb_fst.chrom = both_freqs.chrom AND ahb_fst.pos = both_freqs.pos
JOIN freqs
ON  both_freqs.chrom = freqs.chrom  AND both_freqs.pos = freqs.pos AND pop = "old"
ORDER BY chrom,pos
);
"""
cursor.execute(sql)
```

```python
sql = """
SELECT old.chrom, old.pos, (modern.freq - old.freq) AS diff FROM temp.modern
JOIN temp.old
ON old.chrom = modern.chrom AND old.pos = modern.pos;
"""
cursor.execute(sql)
results = cursor.fetchall()
```

```python
db.close()
```

```python
diffs = pd.DataFrame(np.array(list(results), dtype=[('chrom','a7'),('pos',np.uint),('diff', float)]))
```

## plot data

```python
chr_diagram = BasicChromosome.Organism()
chr_diagram.page_size = (10*cm, 21*cm) #A4 landscape
```

```python
chromlen = {'Group11': 14726556, 'Group10': 12965953, 'Group13': 10288499, 'Group12': 11902654, 'Group15': 10167229, 'Group14': 10253655, 'Group16': 7207165, 'Group5': 14363272, 'Group4': 12718334, 'Group7': 13219345, 'Group6': 18472937, 'Group1': 29893408, 'Group3': 13234341, 'Group2': 15549267, 'Group9': 11120453, 'Group8': 13546544}

for chrom in set(diffs['chrom']):
    features = []
    current_chromosome = diffs[diffs['chrom'] == chrom]
    for line_num in range(len(current_chromosome)):
        f = SeqFeature(FeatureLocation(current_chromosome['pos'][line_num],current_chromosome['pos'][line_num]+1,strand=None))
        f.qualifiers["color"] = 2
        features.append(f)
    body = BasicChromosome.AnnotatedChromosomeSegment(chromlen[chrom], features)
    body.scale = chromlen[chrom]
    cur_chromosome.add(body)
    chr_diagram.add(cur_chromosome)
```

```python
chr_diagram.draw("fst.pdf", "Apis mellifera")
```
